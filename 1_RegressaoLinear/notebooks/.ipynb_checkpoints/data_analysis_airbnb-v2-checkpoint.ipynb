### Categorização inicial de colunas por tipo de dados 

A primeira vista, foi possível identificar os seguintes tipos de colunas:
- tipo texto: 'name','summary', 'space', 'description', 'experiences_offered' e 'neighborhood_overview';
- tipo numérico: 'minimum_minimum_nights', 'maximum_minimum_nights','minimum_maximum_nights','maximum_maximum_nights', 'minimum_nights_avg_ntm',
'maximum_nights_avg_ntm','number_of_reviews_ltm','calculated_host_listings_count_entire_homes','calculated_host_listings_count_private_rooms',
'calculated_host_listings_count_shared_rooms','host_listings_count', 'latitude', 'longitude', 'accommodates','bathrooms','bedrooms', 'beds', 'extra_people', 'minimum_nights', 'number_of_reviews' e 'price';
- tipo booleano: "host_is_superhost", "instant_bookable";
- tipo categóricas: "property_type","room_type", cancellation_policy;
- tipo lista: "amenities"

As colunas restantes, 'listing_url', 'scrape_id' e 'experiences_offered', serão removidas da análise por não apresentarem informação útil para a regressão linear proposta.

```python
#Colunas a serem removidas
columns_to_drop = ['id', 'listing_url', 'scrape_id', 'experiences_offered', 'last_scraped']


#Separando os tipos de variáveis por meio de analise qualitativa
text_columns = ['name','summary', 'space', 'description', 'neighborhood_overview']

numerical_variables = ['minimum_minimum_nights', 'maximum_minimum_nights','minimum_maximum_nights',
                       'maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm',
                       'number_of_reviews_ltm', 'calculated_host_listings_count_entire_homes',
                       'calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms',
                       'host_listings_count', 'latitude', 'longitude', 'accommodates','bathrooms',
                       'bedrooms', 'beds', 'extra_people', 'minimum_nights','number_of_reviews']

boolean_variables = ["host_is_superhost", "instant_bookable"]

categoric_variables = ["property_type","room_type", "cancellation_policy"]
```

Devido ao grande número de dados, apenas removerei as linhas que possuirem dados faltantes em df_train.

```python
df_train.isnull().sum()
```

```python
#Removendo linhas com valores faltantes
df_train.dropna(inplace=True)
```

Dentre as colunas numéricas, algumas necessitam de um tratamento especial. Primeiramente, as colunas as colunas 'extra_people' e 'price' que têm carater numérico mas nesse caso aparece como texto devido ao emprego da cotação em dolar. 

```python
def format_cash(value):
    formated_value = re.sub(r'[\$\,]','',value)
    return float(formated_value)

#Formatando colunas com formato de moeda
df_train.loc[:,["extra_people", "price"]] = df_train.loc[:,["extra_people", "price"]].map(format_cash)
df_test.loc[:,["extra_people"]] = df_test.loc[:,["extra_people"]].map(format_cash)
```

Para o caso das colunas tipo texto, também as removerei devido as multiplas formas de enviesamento que podem ocorrer nesse tipo de dado. 

```python
#Removendo as colunas sem valor sem informação útil para as amostras de treino e de teste
df_train = df_train.drop(columns=columns_to_drop+text_columns)
df_test = df_test.drop(columns=columns_to_drop+text_columns)
```

#### Remoção de Outliers

```python
df_train[numerical_variables+["price"]].describe()
```

Verificando a distribuição de preços de airbnbs

```python
plt.figure(figsize=(10,8))
sns.histplot(df_train["price"])
plt.title('Distribuição de Precos Antes da Remoção de Outliers')
plt.show()
```

Pela distribuição de preços, vemos existem outliers com valores de preço muito acima dos valores comuns. Para tratar desse fato, utilizarei os intervalo do interquartile para definir quais valores devem permanecer. 

```python
# Calculando Q1 (25th percentil) e Q3 (75th percentil)
Q1 = df_train['price'].quantile(0.25)
Q3 = df_train['price'].quantile(0.75)

# Calculando IQR
IQR = Q3 - Q1

# Definindo limite superior para os outliers
upper_bound = Q3 + 1.5 * IQR

# Filtrando DataFrame
df_train = df_train[(df_train['price'] <= upper_bound)]
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df_train["price"])
plt.title('Distribuição de Precos Após da Remoção de Outliers')
plt.show()
```

### Tratamento das colunas por tipo de dados

Para manter o número de registros nos dados de teste, tratarei os valores nulos de modo diferente.

```python
df_test.isnull().sum()
```

```python
df_test.columns
```

```python
#Colunas para preencher valores vazios com a média
columns_to_fill_with_the_mean = ['minimum_nights_avg_ntm', 'maximum_nights_avg_ntm']

#Colunas para preencher valores faltantes com a média arredondada
columns_to_fill_with_rounded_mean = ['minimum_minimum_nights', 'maximum_minimum_nights',
       'minimum_maximum_nights', 'maximum_maximum_nights',
       'number_of_reviews_ltm', 'calculated_host_listings_count_entire_homes',
       'calculated_host_listings_count_private_rooms',
       'calculated_host_listings_count_shared_rooms',
        'bathrooms', 'bedrooms', 'beds', 'host_listings_count']

#Preenchendo valores faltantes
for col in columns_to_fill_with_the_mean:
    mean = df_test[col].mean()
    df_test[col] = df_test[col].fillna(value=mean)

for col in columns_to_fill_with_rounded_mean:
    rounded_mean = round(df_test[col].mean())
    df_test[col] = df_test[col].fillna(value=rounded_mean)

#Utilizando backward fill para preencher os valores faltantes da coluna host_is_superhost
df_test["host_is_superhost"] = df_test["host_is_superhost"].bfill()
```

#### Colunas tipo numéricas

```python
#Garantindo que todas as colunas restantes estão como tipo float
df_train[numerical_variables] = df_train[numerical_variables].astype(float)
df_test[numerical_variables] = df_test[numerical_variables].astype(float)
```

Para tratar as colunas vazias numéricas, apenas utilizarei aplicarei a média dos valores já existentes.

#### Colunas tipo booleano

Nesse caso, basta aplicar os valores 0 e 1 para os valores f e t, respectivamente:

```python
#Formatando colunas tipo booleano
d = {'f':0, 't':1}
df_train.loc[:,boolean_variables] = df_train.loc[:,boolean_variables].map(lambda x: d.get(x)).astype(int)
df_test.loc[:,boolean_variables] = df_test.loc[:,boolean_variables].map(lambda x: d.get(x)).astype(int)
```

#### Colunas categóricas

Para as colunas categóricas, utilizarei o encoding de variáveis dummies

```python
#Concatenando Dataframes para a realização do encoding de modo uniforme entre os dois dados:
unified_df = pd.concat([df_train, df_test])

#Transformando colunas categóricas em dummies
unified_df = pd.get_dummies(data=unified_df, columns= categoric_variables).reset_index(drop=True)

# Separando amostras de treino e de teste novamente
df_train_encoded = unified_df.loc[:len(df_train)-1, : ].copy().reset_index(drop=True)
df_test_encoded = unified_df.loc[len(df_train):, : ].copy().reset_index(drop=True)
```

#### Coluna tipo lista

A coluna "amenities" é do tipo lista e parece ter informações bastante relevantes com relação ao preço do airbnb.

```python
# Função para formatar os dados da coluna amenities de modo a se obter uma lista com cada um dos itens separados.
def format_amenities(string):
    splitted = re.sub(r'{|}|"', "", string).split(",")
    return splitted
```

```python
# Aplicando a formatação à coluna amenities
df_train_encoded["amenities"] = df_train_encoded["amenities"].apply(format_amenities)
df_test_encoded["amenities"] = df_test_encoded["amenities"].apply(format_amenities)
```

```python
#Verificando o número de valores diferentes existentes na coluna amenities
unique_values_amenities = []

#Obtendo todos os possíveis itens em amenities para df_train
for row in df_train_encoded["amenities"].values:
    for item in row:
        if item not in unique_values_amenities:
            unique_values_amenities.append(item)

#Adicionando todos os possíveis itens em amenities em df_test
for row in df_test_encoded["amenities"].values:
    for item in row:
        if item not in unique_values_amenities:
            unique_values_amenities.append(item)
```

```python
#Número de valores únicos em amenities
len(unique_values_amenities)
```

Devida a enorme quantidade de valores diferentes que podem existir nessa coluna, apenas levarei em conta o número de acomodações em cada airbnb

```python
df_train_encoded["amenities"] = df_train_encoded["amenities"].apply(lambda x: len(x))
df_test_encoded["amenities"] = df_test_encoded["amenities"].apply(lambda x: len(x))
```

### Treinamento do Modelo

#### Separando entradas e saídas de treinamento e teste

```python
X_train = df_train_encoded.loc[:,df_train_encoded.columns != 'price'].copy()
Y_train = df_train_encoded.loc[:,['price']].copy()
X_test = df_test_encoded.loc[:,df_test_encoded.columns != 'price'].copy()
```

#### Treinando Modelo

```python
#Importando modelo de Regressão Linear
from sklearn.linear_model import LinearRegression
```

```python
#Realizando o treinamento do modelo
reg = LinearRegression().fit(X_train, Y_train)
```

```python
#Avaliando score
reg.score(X_train, Y_train)
```

```python
predictions = reg.predict(X_test)
```
